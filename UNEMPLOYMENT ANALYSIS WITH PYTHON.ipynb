# UNEMPLOYMENT ANALYSIS WITH PYTHON

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading dataset

```python
# Load the dataset
data = pd.read_csv("C:/B.Tech/V SEM/OASIS internship/Task-2/archive (9)/Unemployment in India.csv")
```

## EDA

```python
data.info()
```

```python
# Print the column names to see the actual column names in your dataset
print(data.columns)
```

```python
# Convert the 'Date' column to a datetime object
data['  Date '] = pd.to_datetime(data['  Date '])
```

```python
# Summary statistics
data.describe()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Data distribution and histograms
plt.figure(figsize=(12, 6))
sns.histplot(data[' Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.title('Distribution of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the average unemployment rate
average_unemployment_rate = data[' Estimated Unemployment Rate (%)'].mean()
average_unemployment_rate
```

```python
# Plot the unemployment rate over time
plt.figure(figsize=(12, 6))
plt.plot(data[' Date'], data[' Estimated Unemployment Rate (%)'], label='Unemployment Rate')
plt.axhline(average_unemployment_rate, color='red', linestyle='--', label='Average Unemployment Rate')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```
